#### 1a) What are the causes of over fitting in deep learning model? name two method which can be used to resolve the over fitting issue in such architecture.

##### Causes of Overfitting in Deep Learning Models:
1. Complex Model Architecture: A model with too many parameters (layers or neurons) can memorize the training data rather than generalize to unseen data.
2. Insufficient Training Data: When the training dataset is too small, the model might learn patterns specific to the training set instead of generalizing to other data.

##### Methods to Resolve Overfitting:
1. Regularization:
    - L1/L2 Regularization: Adds a penalty term to the loss function to constrain the model weights, reducing the risk of overfitting.
    - Dropout: Randomly disables a fraction of neurons during training, forcing the model to learn more robust features.
2. Data Augmentation:
    - Apply transformations to the training data (e.g., rotations, flips, cropping, or scaling) to artificially increase the dataset size and diversity, improving the model's generalization.

These methods are widely used in deep learning to tackle overfitting effectively.

#### 1b) what is the usage of convolution layer, pooling layer and dense layer in convolution neural network(CNN) architecture?

In a Convolutional Neural Network (CNN) architecture, the Convolutional Layer, Pooling Layer, and Dense Layer each have specific roles and functions. Here's an explanation:

1. Convolutional Layer
    - Usage: Extracts features from input data (e.g., images).
    - How It Works: Applies convolutional filters (kernels) to the input, which slide across the input image to compute feature maps.
    - Purpose:
        - Captures spatial and hierarchical patterns in data (e.g., edges, textures, and objects in an image).
        - Preserves spatial relationships between pixels.
    - Output: Produces feature maps highlighting important patterns.
2. Pooling Layer
    - Usage: Reduces the spatial dimensions of feature maps (downsampling).
    - How It Works: Aggregates values in a neighborhood (e.g., using Max Pooling or Average Pooling) to produce a smaller output.
    - Purpose:
        - Reduces computational cost and memory usage by decreasing dimensions.
        - Retains the most significant features.
        - Adds invariance to small translations or distortions in the input.
    - Output: Smaller feature maps with retained essential features.
3. Dense Layer (Fully Connected Layer)
    - Usage: Performs high-level reasoning and classification based on extracted features.
    - How It Works: Connects every neuron in the current layer to every neuron in the next layer, processing the flattened feature maps from previous layers.
    - Purpose:
        - Combines all extracted features to form a final output, such as class probabilities in classification tasks.
        - Learns non-linear combinations of features for decision-making.
    - Output: Usually produces the final predictions or classifications.

##### Summary of Roles:
|Layer	| Purpose |	Example|
|-------|---------|--------|
|Convolutional |	Feature extraction |	Detects edges, textures, or shapes.|
|Pooling |	Dimensionality reduction |	Downsamples feature maps.|
|Dense |	Decision-making and classification |	Maps features to output labels or predictions.|

These layers work together to process input data hierarchically, from low-level feature extraction to high-level decision-making. 

#### 1c) What is the vanishing gradient problem in deep learning network? How to get rid of this?

##### Vanishing Gradient Problem in Deep Learning
The vanishing gradient problem occurs during training deep neural networks when the gradients of the loss function with respect to the network's weights become very small. This issue arises primarily in networks with many layers, especially in recurrent neural networks (RNNs) and deep feedforward networks.

##### Causes of the Vanishing Gradient Problem
1. Sigmoid and Tanh Activation Functions:
    - These activation functions compress input values into small ranges:
        - Sigmoid: [0, 1]
        - Tanh: [-1, 1]
    - The derivative of these functions becomes very small for large input values, leading to negligible gradients during backpropagation.
2. Chain Rule of Backpropagation:
    - During backpropagation, gradients are computed as a product of derivatives layer-by-layer. If the derivatives are less than 1, the gradient diminishes exponentially as it propagates back through layers.

##### Effects of the Vanishing Gradient Problem
- Slow Learning: Early layers learn extremely slowly or stop learning altogether.
- Poor Model Performance: The network struggles to optimize weights effectively, leading to suboptimal results.

##### Ways to Overcome the Vanishing Gradient Problem
1. Activation Functions:
    - ReLU (Rectified Linear Unit): ReLU activation does not squash values, ensuring gradients do not vanish (though care is needed for dead neurons).
    - Variants of ReLU: Leaky ReLU, Parametric ReLU, or Exponential Linear Unit (ELU) further mitigate vanishing gradients.
2. Weight Initialization:
    - Use advanced initialization methods like Xavier Initialization or He Initialization to ensure weights start in a range that avoids extreme shrinking of gradients.
3. Batch Normalization:
    - Normalizes the input to each layer, maintaining mean and variance, which helps in preserving gradients.
4. Gradient Clipping:
    - Restricts gradients within a specified range to prevent extreme shrinking or exploding.
5. Residual Connections (ResNets):
    - Introduce skip connections, allowing gradients to flow directly to earlier layers, bypassing intermediate layers.
6. Use Smaller Learning Rates:
    - A smaller learning rate can help gradients stabilize, although this alone may not fully resolve the issue.

##### Key Takeaways
- The vanishing gradient problem is a major bottleneck in training deep networks.
- Modern techniques like ReLU activations, batch normalization, and residual connections are effective solutions.
- These solutions collectively enable the successful training of deep architectures like ResNets and transformers.

#### 1d) List down various performance matrices used for object detection. What is the use of non-maximal suppression in object detection algorithm?

##### Performance Metrics for Object Detection
Object detection performance is evaluated using various metrics to assess the accuracy and efficiency of the model. Commonly used metrics include:

1. Intersection over Union (IoU):
    - Measures the overlap between the predicted bounding box and the ground truth box.
    - Formula: 𝐼𝑜𝑈 = Area of Overlap / Area of Union
    - A higher IoU indicates better localization.
2. Mean Average Precision (mAP):
    - Combines precision and recall over different IoU thresholds.
    - Average precision (AP) is calculated for each class, and mAP is the mean across all classes.
    - Commonly used IoU thresholds: 0.5 (mAP@50) or 0.5:0.95 (COCO mAP).
3. Precision:
    - Measures the proportion of correctly predicted positive instances.
    - Formula: Precision = True Positives / True Positives + False Positives
4. Recall:
    - Measures the proportion of actual positives correctly predicted.
    - Formula: Recall = True Positives / True Positives + False Negatives
5. F1 Score:
    - Harmonic mean of precision and recall, used for balanced evaluation.
    - Formula: 𝐹1 = 2 × Precision × Recall / Precision + Recall
6. Frames per Second (FPS):
    - Evaluates the inference speed of the object detection model.
    - Important for real-time applications.
7. Log Average Miss Rate (LAMR):
    - Combines false positives per image (FPPI) and miss rate into a single metric.
    - Common in pedestrian detection tasks.
8. Detection Rate:
    - Measures the proportion of correctly detected objects out of all ground truth objects.

##### Non-Maximal Suppression (NMS) in Object Detection

    Purpose:
        - Non-Maximal Suppression is used to remove redundant or overlapping bounding boxes for the same object, keeping only the most confident prediction.
        
    How It Works:
        1. Score Ranking: Rank all predicted bounding boxes based on their confidence scores.
        2. Thresholding: Remove boxes with confidence scores below a set threshold.
        3. IoU Comparison:
            - For each remaining box, calculate the IoU with other boxes.
            - Suppress (discard) boxes that overlap significantly (IoU > threshold).
        4. Retain: Keep the box with the highest confidence score among overlapping boxes.
    
    Benefits:
        - Reduces duplicate detections of the same object.
        - Improves clarity and accuracy of the detection output.
        
##### Summary
    - Performance metrics like IoU, mAP, precision, recall, and FPS assess object detection models.
    - Non-Maximal Suppression is essential for post-processing to ensure clear and accurate object detection results.


#### 1e) What is the requirement of an activation function in deep neural network architecture? Define the following activation functions and their usage, Tanh(), ReLU() and Softmax().

##### Requirement of an Activation Function in Deep Neural Networks
- Non-linearity: Activation functions introduce non-linear transformations to the output of neurons, enabling the network to learn complex patterns and representations in data. Without non-linearity, a neural network would behave like a linear model regardless of the number of layers.
- Decision Boundaries: They help the network learn and create complex decision boundaries for classification and regression tasks.
- Gradient Propagation: Activation functions ensure gradients flow effectively during backpropagation, which is critical for learning.

##### Activation Functions and Their Usage
1. Tanh (Hyperbolic Tangent)
- Definition:

Tanh(x)= e^x - e^ −x / e^x + e^−x

    - Maps input values to the range [-1, 1].

- Usage:
    - Suitable for hidden layers where the input may have both positive and negative values.
    - Helps center the data around zero, making optimization easier.
- Advantages:
    - Zero-centered output improves gradient flow.
    - Works well for shallow networks.
- Disadvantages:
    - Suffers from the vanishing gradient problem for large or small input values, as gradients approach zero.

2. ReLU (Rectified Linear Unit)
- Definition:
        ReLU(x)=max(0,x)
    - Outputs zero for negative inputs and the input value itself for positive inputs.
- Usage:
    - Most commonly used in hidden layers of deep neural networks due to simplicity and efficiency.
    - Activates only some neurons at a time, making the network sparse and reducing computation.
- Advantages:
    - Computationally efficient (simple comparison and output).
    - Does not saturate for large input values (avoids vanishing gradients).
- Disadvantages:
    - Can lead to dead neurons if a neuron outputs zero consistently and never recovers (gradient is zero for negative inputs).
    - Variants like Leaky ReLU or Parametric ReLU address this limitation.
 
3. Softmax
- Definition:

    Softmax(Zi)= e^Xi / ∑(j=1, K) e^Zj
    
    - Converts a vector of raw scores Zi into probabilities, where K is the number of classes, and the probabilities sum to 1.
- Usage:
    - Used in the output layer for multi-class classification problems.
    - Each output represents the probability of belonging to a specific class.
- Advantages:
    - Provides normalized probabilities for multi-class problems.
- Disadvantages:
    - Can amplify small differences in input values, making it sensitive to outliers.
    - Not ideal for binary classification (use sigmoid instead).

##### Summary of Activation Functions
|Activation|	Output Range|	Use Case|	Pros|	Cons|
|----------|----------------|-----------|-------|------|
|Tanh|	[-1, 1]|	Hidden layers (zero-centered data)|	Zero-centered output|	Vanishing gradient problem|
|ReLU|	[0, ∞)|	Hidden layers (deep networks)|	Efficient, avoids vanishing gradient|	Dead neurons|
|Softmax|	[0, 1] (sums to 1)|	Output layer (multi-class problems)|	Normalized probabilities|	Sensitive to outliers|


#### 2) Build convolution neural network to classify the six different types of food types.
#### Dataset_Folder and Name:

Conditions to consider:
- Parameters should not cross 20000
- Should not use more than four layers(except input and output, including convolution and dense layers).
- Use adam optimizer.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define parameters
input_shape = (128, 128, 3)  # Assuming images are resized to 128x128 with 3 channels
num_classes = 6              # Number of food types

# Create the model
model = Sequential([
    # Convolutional layer 1
    Conv2D(16, (3, 3), activation='relu', input_shape=input_shape),
    MaxPooling2D((2, 2)),

    # Convolutional layer 2
    Conv2D(32, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Flatten layer
    Flatten(),

    # Dense layer
    Dense(64, activation='relu'),

    # Output layer
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer=Adam(),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Summary to verify parameters
model.summary()

# Data preprocessing
train_datagen = ImageDataGenerator(rescale=1.0/255, validation_split=0.2)

train_generator = train_datagen.flow_from_directory(
    '3_food_classes',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    '3_food_classes',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

print(train_generator)


# Train the model
epochs = 10
model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator
)

# Save the model
model.save('food_classification_cnn.h5')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Define parameters
input_shape = (128, 128, 3)  # Assuming images are resized to 128x128 with 3 channels
num_classes = 3              # Number of food types

# Create the model
model = Sequential([
    # Convolutional layer 1 with Batch Normalization
    Conv2D(16, (3, 3), activation='relu', input_shape=input_shape),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    # Convolutional layer 2 with Batch Normalization
    Conv2D(32, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    # Convolutional layer 3 with Batch Normalization
    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    # Flatten layer
    Flatten(),

    # Dense layer with Dropout
    Dense(128, activation='relu'),
    Dropout(0.5),

    # Output layer
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer=RMSprop(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Summary to verify parameters
model.summary()

# Data augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    'Dataset_Folder',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    'Dataset_Folder',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Define callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('improved_food_classification_model.h5', save_best_only=True, monitor='val_loss')

# Train the model
epochs = 20
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=validation_generator,
    callbacks=[early_stopping, model_checkpoint]
)

# Save the weights
model.save_weights('improved_model_weights.h5')
```

4. Optimizer: Used RMSprop for better convergence.

    - Early stopping prevents overfitting and stops training when the validation loss doesn't improve.

```python
import numpy as np
import os
import cv2
from sklearn.model_selection import train_test_split
import tensorflow as tf
import segmentation_models as sm

# Define paths
data_path = 'Unet_Dataset/images/'
mask_path = 'Unet_Dataset/masks/'

# Load images and masks
image_list = []
mask_list = []
image_size = (128, 128)

for img_file in sorted(os.listdir(data_path)):
    img = cv2.imread(os.path.join(data_path, img_file), cv2.IMREAD_GRAYSCALE)  # Load grayscale images
    img = cv2.resize(img, image_size)
    image_list.append(img)

for mask_file in sorted(os.listdir(mask_path)):
    mask = cv2.imread(os.path.join(mask_path, mask_file), cv2.IMREAD_GRAYSCALE)
    mask = cv2.resize(mask, image_size)
    mask_list.append(mask)

# Convert to numpy arrays
images = np.array(image_list)  # Shape: (96, 128, 128)
masks = np.array(mask_list)    # Shape: (96, 128, 128)

# Expand dimensions to match model input
images = np.expand_dims(images, axis=-1)  # Shape: (96, 128, 128, 1)
masks = np.expand_dims(masks, axis=-1)    # Shape: (96, 128, 128, 1)

# Normalize images and masks
images = images / 255.0  # Scale images to [0, 1]
masks = masks / 255.0    # Scale masks to [0, 1]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(images, masks, test_size=0.2, random_state=42)

# Define UNet model
model = sm.Unet(
    backbone_name='resnet34',  # Pretrained UNet backbone
    encoder_weights=None,     # Use None if no internet access
    input_shape=(128, 128, 1),
    classes=1,                # Binary segmentation
    activation='sigmoid'      # Sigmoid activation for binary masks
)

# Compile the model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=sm.losses.binary_crossentropy,  # Binary crossentropy loss
    metrics=[sm.metrics.iou_score]       # IoU as metric
)

# Define callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)

# Train the model
history = model.fit(
    X_train, y_train,
    epochs=2,
    batch_size=2,  # Reduce batch size for memory efficiency
    validation_data=(X_test, y_test),
    callbacks=[early_stopping]
)

# Evaluate the model
loss, iou = model.evaluate(X_test, y_test, batch_size=2)
print(f"Test Loss: {loss:.4f}, Test IoU: {iou:.4f}")
```

##### Key Steps in the Code:
1. Data Loading and Preprocessing:
    - Images and masks are loaded and resized to 128x128.
    - Images are normalized to [0, 1].
    - Masks are normalized and reshaped to include a channel dimension.
2. Data Splitting:
    - The dataset is split into training and testing sets with an 80-20 ratio.
3. Model Definition:
    - The UNet model is created using a resnet34 backbone from the segmentation_models library.
    - Encoder weights are set to None for offline training.
4. Compilation:
    - Binary cross-entropy is used as the loss function.
    - Intersection over Union (IoU) is used as the evaluation metric.
5. Training:
    - Trains the model for 2 epochs using a small batch size of 2 for memory efficiency.
    - Implements early stopping to prevent overfitting.
6. Evaluation:
    - Evaluates the model on the test set and prints the loss and IoU score.
